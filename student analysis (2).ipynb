```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Performance_Analysis.csv")

df.head(4)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Average_Score"] = (
    df["Math_Score"] + 
    df["Science_Score"] + 
    df["English_Score"]
) / 3
```

```python
top_students = df.sort_values(by="Average_Score", ascending=False).head(8)
("Top 5 Students:")
(top_students[["Name", "Average_Score"]])
```

```python
gender_avg = df.groupby("Gender")["Average_Score"].mean()
("Average Score by Gender:")
(gender_avg)
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Study_Hours_per_Week"], df["Average_Score"])
plt.xlabel("Study Hours per Week")
plt.ylabel("Average Score")
plt.title("Study Hours vs Average Score")
plt.show()
```

```python
plt.figure()
plt.bar(df["Attendance_Percentage"], df["Average_Score"])
plt.xlabel("Attendance Percentage")
plt.ylabel("Average Score")
plt.title("Attendance vs Average Score")
plt.show()
```
